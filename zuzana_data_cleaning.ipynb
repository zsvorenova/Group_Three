```python
spotify_grouped_track = pd.DataFrame(spotify_df["track_id"].value_counts())
spotify_grouped_track["avg_popularity"] = spotify_df.groupby("track_id")["popularity"].mean()
spotify_grouped_track["min_popularity"] = spotify_df.groupby("track_id")["popularity"].min()
spotify_grouped_track["max_popularity"] = spotify_df.groupby("track_id")["popularity"].max()
spotify_grouped_track["diff_popularity"] = spotify_grouped_track["max_popularity"] - spotify_grouped_track["min_popularity"]
#spotify_grouped_track["mode_popularity"] = spotify_df.groupby("track_id")["popularity"].mode(axis=1)
spotify_grouped_track["no_issues"] = spotify_grouped_track["min_popularity"] == spotify_grouped_track["max_popularity"]

#https://stackoverflow.com/questions/52336392/convert-boolean-to-string-in-dataframe
# steps to convert booleans to strings so we can filter later on it too
#mask = spotify_grouped_track.applymap(type) != bool
#d = {True: "TRUE", False: "FALSE"}
#spotify_grouped_track = spotify_grouped_track.where(mask, spotify_grouped_track.replace(d))

multiple_tracks = spotify_grouped_track.loc[spotify_grouped_track["track_id"] > 1]
multiple_tracks = spotify_grouped_track.loc[spotify_grouped_track["no_issues"] == False]
# can we use loc in one line for 2 different columns?
multiple_tracks
```

```python
multiple_tracks_sort = multiple_tracks.sort_values(["diff_popularity"], ascending = False)
multiple_tracks_sort
```
