```python
import spotipy
from zuzana_config import Client_ID, Client_Secret
import pandas as pd
import json
from pprint import pprint
```

```python
token = spotipy.oauth2.SpotifyClientCredentials(client_id=Client_ID, client_secret=Client_Secret)
cache_token = token.get_access_token()
sp = spotipy.Spotify(cache_token)
```

```python
# get information on playlist: Today's Top Hits by Spotify
username = "spotify"
playlist_id = "37i9dQZF1DXcBWIGoYBM5M"
playlist_name = sp.playlist(playlist_id)['name'].replace(" ","_")
playlist_name
```

```python
playlist = []
playlist = sp.user_playlist_tracks(username, playlist_id)["items"]

artist = []
album = []
track_name = []
track_id = []
popularity = []

for track in playlist:
    artist.append(track["track"]["album"]["artists"][0]["name"])
    album.append(track["track"]["album"]["name"])
    track_name.append(track["track"]["name"])
    track_id.append(track["track"]["id"]) 
    popularity.append(track["track"]["popularity"])

playlist_data = pd.DataFrame({'artist_name' : artist, 'album': album, 'track_name' : track_name, 'track_id' : track_id, 'popularity' : popularity})
playlist_data

output_file = playlist_name.replace("'","") + ".csv"
output_file

playlist_data.to_csv(output_file)
```
